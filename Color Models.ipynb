```python
import cv2
import numpy as np
```

```python
#Creating an empty function which does nothing
def do_nothing(x):
    pass

#Creating a black image
img = np.zeros((300,512,3),np.uint8)
cv2.namedWindow('Image')

#Creating trackbars for RGB values
cv2.createTrackbar('Red','Image',0,255,do_nothing)
cv2.createTrackbar('Green','Image',0,255,do_nothing)
cv2.createTrackbar('Blue','Image',0,255,do_nothing)

switch = '0 : OFF \n1 : ON'
cv2.createTrackbar(switch, 'image',0,1,do_nothing)

while True:
    cv2.imshow('Image',img)
    key = cv2.waitKey(10) & 0xFF
    if key == 27:
        break
    
    #Getting trackbar positions of all the trackbars
    r = cv2.getTrackbarPos('Red','Image')
    g = cv2.getTrackbarPos('Green','Image')
    b = cv2.getTrackbarPos('Blue','Image')
    s = cv2.getTrackbarPos(switch,'Image')
    
    if s == 0:
        img[:] = 0
    else:
        img[:] = [b,g,r]
        
cv2.destroyAllWindows()
    
```

```python
img2 = np.zeros((300,512,3),np.uint8)
cv2.namedWindow('HSV-Image')

#Creating trackbars for HSV values
cv2.createTrackbar('Hue (째)','HSV-Image',0,240,do_nothing)
cv2.createTrackbar('Saturation','HSV-Image',0,100,do_nothing)
cv2.createTrackbar('Value','HSV-Image',0,100,do_nothing)

switch = '0 : OFF \n1 : ON'
cv2.createTrackbar(switch, 'HSV-Image',0,1,do_nothing)

while True:
    
    img2 = cv2.cvtColor(img2,cv2.COLOR_BGR2HSV)
    cv2.imshow('HSV-Image',img2)
    key = cv2.waitKey(10) & 0xFF
    if key == 27:
        break
        
    h = cv2.getTrackbarPos('Hue (째)','HSV-Image')
    sat = cv2.getTrackbarPos('Saturation','HSV-Image')
    v = cv2.getTrackbarPos('Value','HSV-Image')
    s = cv2.getTrackbarPos(switch,'HSV-Image')
    
    if s == 0:
        img2[:] = 0
    else:
        img2[:] = [h,s,v]
        
cv2.destroyAllWindows()
    
```
